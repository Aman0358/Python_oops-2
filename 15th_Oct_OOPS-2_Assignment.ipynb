```python
# Inheritance:
# What is inheritance in Python? Explain its significance in object-oriented programming.
# Inheritance is a fundamental concept in object-oriented programming (OOP), and it plays a crucial role in Python as well as in many other OOP languages. It allows you to create a new class that is a modified version of an existing class. Inheritance is a way to achieve code reuse and to model the "is-a" relationship between classes.
# In Python, inheritance works as follows:
# Base Class (Parent Class): The class that you want to inherit from is called the base class or the parent class. It defines the common attributes and methods that are shared among multiple related classes.
# Derived Class (Child Class): The class that inherits from the base class is called the derived class or child class. It inherits the attributes and methods of the base class and can also have its own attributes and methods. The derived class can add new functionality or modify the behavior of the base class.
# Here's an example to illustrate inheritance in Python:
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        pass

class Dog(Animal):
    def speak(self):
        return f"{self.name} says Woof!"

class Cat(Animal):
    def speak(self):
        return f"{self.name} says Meow!"

dog = Dog("Buddy")
cat = Cat("Whiskers")

print(dog.speak())  # Output: Buddy says Woof!
print(cat.speak())  # Output: Whiskers says Meow!

```

```python
# Differentiate between single inheritance and multiple inheritance in Python. Provide examples for each.
# Single Inheritance:

# In single inheritance, a class can inherit from only one base class.
# It promotes a simpler and more linear class hierarchy, making it easier to understand and maintain.
# It is the default and most common form of inheritance in Python.
# Example of Single Inheritance:
class Parent:
    def show(self):
        print("This is the Parent class")

class Child(Parent):
    def display(self):
        print("This is the Child class")

child = Child()
child.show()  # Accessing the show method from the Parent class
child.display()  # Accessing the display method from the Child class

```

```python
# Multiple Inheritance:

# In multiple inheritance, a class can inherit from multiple base classes. This means a class can have attributes and methods from more than one parent class.
# It provides greater flexibility but can lead to complex and ambiguous class hierarchies if not used carefully.
# Python supports multiple inheritance, but it requires careful consideration to avoid conflicts and ambiguities, which can be resolved using method resolution order (MRO).
# Example of Multiple Inheritance:
class Parent1:
    def show(self):
        print("This is Parent1 class")

class Parent2:
    def display(self):
        print("This is Parent2 class")

class Child(Parent1, Parent2):
    def print_all(self):
        self.show()  # Accessing the show method from Parent1
        self.display()  # Accessing the display method from Parent2

child = Child()
child.print_all()  # Accessing methods from both Parent1 and Parent2
```

```python
# 3. How can you achieve encapsulation in Python classes? Provide an example.
class Student:
    def __init__(self, name, age):
        self._name = name  # Private attribute with a leading underscore
        self._age = age    # Private attribute with a leading underscore

    # Getter method for name
    def get_name(self):
        return self._name

    # Setter method for name
    def set_name(self, name):
        if isinstance(name, str):
            self._name = name
        else:
            print("Name should be a string.")

    # Getter method for age
    def get_age(self):
        return self._age

    # Setter method for age
    def set_age(self, age):
        if age >= 0:
            self._age = age
        else:
            print("Age should be a non-negative value.")

    # Property for name
    @property
    def name(self):
        return self._name

    # Property setter for name
    @name.setter
    def name(self, name):
        if isinstance(name, str):
            self._name = name
        else:
            print("Name should be a string.")

    # Property for age
    @property
    def age(self):
        return self._age

    # Property setter for age
    @age.setter
    def age(self, age):
        if age >= 0:
            self._age = age
        else:
            print("Age should be a non-negative value.")

# Create a Student object
student = Student("Alice", 20)

# Access attributes using getter methods
print(student.get_name())  # Output: Alice
print(student.get_age())   # Output: 20

# Modify attributes using setter methods
student.set_name("Bob")
student.set_age(25)
print(student.get_name())  # Output: Bob
print(student.get_age())   # Output: 25

# Access attributes using properties (no need for explicit getter methods)
print(student.name)  # Output: Bob
print(student.age)   # Output: 25

# Modify attributes using properties (no need for explicit setter methods)
student.name = "Charlie"
student.age = -5  # This will trigger the validation message
print(student.name)  # Output: Charlie
print(student.age)   # Output: 25 (not changed due to validation)
```

```python
# 6. Explain the purpose of getter and setter methods in encapsulation. Provide examples.
class Temperature:
    def __init__(self, value=0.0):
        self._celsius = value  # Private attribute with a leading underscore

    # Getter method for celsius
    def get_celsius(self):
        return self._celsius

    # Setter method for celsius
    def set_celsius(self, value):
        if value < -273.15:  # Absolute zero in Celsius
            print("Temperature cannot be below absolute zero.")
        else:
            self._celsius = value

    # Getter method for fahrenheit
    def get_fahrenheit(self):
        return (self._celsius * 9/5) + 32

    # Setter method for fahrenheit
    def set_fahrenheit(self, value):
        celsius = (value - 32) * 5/9
        self.set_celsius(celsius)  # Use the celsius setter to perform the conversion

# Create a Temperature object
temperature = Temperature(25)

# Access the temperature in Celsius using the getter
celsius = temperature.get_celsius()
print("Celsius:", celsius)  # Output: Celsius: 25

# Modify the temperature in Celsius using the setter
temperature.set_celsius(30)
celsius = temperature.get_celsius()
print("Celsius:", celsius)  # Output: Celsius: 30

# Access the temperature in Fahrenheit using the getter
fahrenheit = temperature.get_fahrenheit()
print("Fahrenheit:", fahrenheit)  # Output: Fahrenheit: 86

# Modify the temperature in Fahrenheit using the setter
temperature.set_fahrenheit(95)
celsius = temperature.get_celsius()
print("Celsius:", celsius)  # Output: Celsius: 35
```

```python
# 7. What is name mangling in Python, and how does it affect encapsulation?

# Name mangling is a mechanism in Python that affects attribute name resolution, particularly for attributes with double underscores (e.g., __attribute). It is used to make the name of a class attribute more unique to avoid naming conflicts with subclasses and is often associated with a form of name "obfuscation." This feature is applied to class attributes to effectively make them "private" in a way that they are still accessible but less likely to be accidentally overridden in subclasses. Here's how it works and how it affects encapsulation:
class MyClass:
    def __init__(self):
        self.__attribute = 42  # Private attribute with double underscores

    def display_attribute(self):
        print(self.__attribute)  # Access within the class

class MySubClass(MyClass):
    def modify_attribute(self):
        self.__attribute = 100  # Shadowing the attribute with a different value

# Create an object of the base class
obj = MyClass()
obj.display_attribute()  # Output: 42

# Create an object of the subclass
sub_obj = MySubClass()
sub_obj.display_attribute()  # Output: 42 (uses the mangled name)

sub_obj.modify_attribute()  # Modifies a different attribute (name not mangled)
sub_obj.display_attribute()  # Output: 42

print(obj.__dict__)  # {'_MyClass__attribute': 42}
print(sub_obj.__dict__)  # {'_MyClass__attribute': 100, '__attribute': 100}
```

```python
# 12. Explain the concept of property decorators in Python and how they relate to encapsulation.
# Property decorators in Python are a mechanism that allows you to control access to class attributes and add additional behavior when getting, setting, or deleting those attributes. They play a significant role in encapsulation, which is one of the fundamental principles of object-oriented programming. Encapsulation is the concept of restricting direct access to an object's internal state (attributes) and providing controlled access through methods. Property decorators help achieve encapsulation by allowing you to hide the implementation details of an attribute while exposing a public interface for interacting with it.

# There are three key property decorators in Python:
# @property: This decorator is used to define a method as a getter for an attribute. It allows you to access the attribute as if it were a regular attribute, without needing to call a method explicitly. This can be useful for exposing an attribute publicly while encapsulating its internal implementation.
# Example:
class MyClass:
    def __init__(self):
        self._value = 0

    @property
    def value(self):
        return self._value

obj = MyClass()
print(obj.value)  # Access the 'value' attribute using the getter method.
# @attribute_name.setter: This decorator is used to define a method as a setter for an attribute. It allows you to add logic and checks when setting the attribute. This ensures that you can control what values are assigned to the attribute and maintain data integrity.
# Example:
class MyClass:
    def __init__(self):
        self._value = 0

    @property
    def value(self):
        return self._value

    @value.setter
    def value(self, new_value):
        if new_value < 0:
            raise ValueError("Value must be non-negative")
        self._value = new_value

obj = MyClass()
obj.value = 42  # Setting the 'value' attribute through the setter method.
# @attribute_name.deleter: This decorator is used to define a method as a deleter for an attribute. It allows you to define behavior for when an attribute is deleted, which can be useful for resource management or cleanup.
# Example:
class MyClass:
    def __init__(self):
        self._value = 0

    @property
    def value(self):
        return self._value

    @value.deleter
    def value(self):
        print("Deleting 'value'")
        del self._value

obj = MyClass()
del obj.value  # Deleting the 'value' attribute using the deleter method.
```

```python
# 15. Describe the concept of operator overloading in Python and how it relates to polymorphism. Provide examples using operators like `+` and `*`.
# Operator overloading is a feature in Python that allows you to define how operators should behave for objects of user-defined classes. It enables you to give custom meaning to standard Python operators (e.g., +, -, *, /, <, ==, etc.) for instances of your own classes. Operator overloading is closely related to polymorphism and enables you to achieve polymorphism with operators.
# Operator Overloading with + (Addition):
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        if isinstance(other, Vector):
            return Vector(self.x + other.x, self.y + other.y)
        else:
            raise TypeError("Unsupported operand type for +")

    def __str__(self):
        return f"Vector({self.x}, {self.y})"

v1 = Vector(1, 2)
v2 = Vector(3, 4)

result = v1 + v2
print(result)  # Output: Vector(4, 6)
```

```python
# Operator Overloading with * (Multiplication):
class ComplexNumber:
    def __init__(self, real, imaginary):
        self.real = real
        self.imaginary = imaginary

    def __mul__(self, other):
        if isinstance(other, ComplexNumber):
            real_part = (self.real * other.real) - (self.imaginary * other.imaginary)
            imaginary_part = (self.real * other.imaginary) + (self.imaginary * other.real)
            return ComplexNumber(real_part, imaginary_part)
        else:
            raise TypeError("Unsupported operand type for *")

    def __str__(self):
        return f"{self.real} + {self.imaginary}i"

c1 = ComplexNumber(1, 2)
c2 = ComplexNumber(2, 3)

result = c1 * c2
print(result)  # Output: 4 + 7i
```

```python
# 16. What is dynamic polymorphism, and how is it achieved in Python?
# Dynamic polymorphism, also known as runtime polymorphism or late binding, is a key concept in object-oriented programming. It allows the selection of a specific method to be executed at runtime, based on the actual type of an object, rather than at compile time. In dynamic polymorphism, you can invoke a method on an object, and the correct implementation of that method is determined at runtime, depending on the actual subclass type of the object.

# Dynamic polymorphism is achieved in Python through method overriding and the use of inheritance. Here's how it works in Python:

# 1. **Method Overriding**: Subclasses can provide their own implementation of a method that is already defined in their parent class. This is known as method overriding. When you invoke the overridden method on an object of the subclass, the subclass's method is executed instead of the parent class's method.

# 2. **Inheritance**: Dynamic polymorphism is closely related to inheritance. The subclasses inherit methods from the parent class, and if a subclass overrides a method, it replaces the parent class's method.

# 3. **Runtime Binding**: When you call a method on an object in Python, the interpreter determines the actual type of the object at runtime and selects the correct method implementation to execute. This is known as runtime binding, and it achieves dynamic polymorphism.

# Here's an example that demonstrates dynamic polymorphism in Python:

class Animal:
    def speak(self):
        print("Some generic animal sound")

class Dog(Animal):
    def speak(self):
        print("Woof!")

class Cat(Animal):
    def speak(self):
        print("Meow!")

def animal_sound(animal):
    animal.speak()

dog = Dog()
cat = Cat()

animal_sound(dog)  # Output: "Woof!"
animal_sound(cat)  # Output: "Meow!"


# In this example, we have a base class `Animal` with a `speak()` method. The `Dog` and `Cat` classes override the `speak()` method with their specific sounds. When we call the `animal_sound()` function on `dog` and `cat` objects, the appropriate `speak()` method is executed based on the actual type of the object, achieving dynamic polymorphism.

# Dynamic polymorphism is a fundamental concept in Python and object-oriented programming, allowing you to write flexible and extensible code that can work with objects of different types in a uniform way.
```

```python
# 19. Explain the role of interfaces and abstract classes in polymorphism, drawing comparisons between them.
# Interfaces and abstract classes are two important concepts in object-oriented programming that play a significant role in achieving polymorphism. Both serve as a means to define a common interface or contract that classes must adhere to, but they have distinct characteristics and purposes.

# **Abstract Classes:**
# 1. **Definition:** An abstract class is a class that cannot be instantiated on its own and is typically used as a blueprint for other classes. It can contain both abstract (unimplemented) and concrete (implemented) methods.

# 2. **Common Interface:** Abstract classes can define a common interface by specifying a set of methods that derived classes must implement. They can also provide default implementations for some methods, allowing derived classes to override only those methods that are specific to their functionality.

# 3. **Inheritance:** Abstract classes encourage inheritance. Derived classes must inherit from the abstract class, and they are required to provide concrete implementations for any abstract methods defined in the abstract class.

# 4. **Use Cases:** Abstract classes are used when you want to create a base class with some shared behavior and ensure that all derived classes implement a specific set of methods.

# **Interfaces:**
# 1. **Definition:** An interface is a pure abstract class that defines a contract of methods that must be implemented by any class that adheres to the interface. Interfaces contain only method declarations, and all methods are abstract and unimplemented.

# 2. **Common Interface:** Interfaces serve as a contract for classes that implement them. Any class that implements an interface must provide concrete implementations for all the methods defined in the interface.

# 3. **Implementation:** Classes can implement multiple interfaces, allowing them to adhere to multiple contracts and provide multiple sets of behavior. This is a form of multiple inheritance.

# 4. **Use Cases:** Interfaces are used when you want to ensure that classes adhere to a specific set of method signatures, without specifying any shared behavior or state. They promote loose coupling and allow for greater flexibility.

# **Comparisons:**

# - **Inheritance vs. Implementation:** Abstract classes are typically associated with inheritance, as derived classes extend the abstract class. Interfaces, on the other hand, are associated with implementation, as classes implement interfaces.

# - **Method Implementations:** Abstract classes can provide concrete method implementations along with abstract methods, while interfaces only contain method declarations with no implementations.

# - **Multiple Inheritance:** While some programming languages support multiple interface implementation, abstract classes often do not support multiple inheritance, meaning a class can only inherit from one abstract class.

# - **Shared Behavior:** Abstract classes are used when you want to share some common behavior among related classes. Interfaces, however, are used when you want to enforce a contract without sharing behavior.

# - **Flexibility:** Interfaces provide more flexibility as a class can implement multiple interfaces, allowing it to adhere to multiple contracts. Abstract classes, due to single inheritance, might be more rigid in this regard.

# In summary, both abstract classes and interfaces play a role in achieving polymorphism by defining a common interface for classes to adhere to. The choice between them depends on whether you want to share behavior or enforce a contract, and the specific requirements of your object-oriented design.

```

```python
# 16. Create a Python class for a computer system, demonstrating abstraction by defining common methods (e.g., `power_on()`, `shutdown()`) in an abstract base class.
from abc import ABC, abstractmethod

class ComputerSystem(ABC):
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
        self.powered_on = False

    @abstractmethod
    def power_on(self):
        pass

    @abstractmethod
    def shutdown(self):
        pass

class Desktop(ComputerSystem):
    def power_on(self):
        if not self.powered_on:
            print(f"{self.brand} {self.model} desktop is powering on.")
            self.powered_on = True
        else:
            print("Desktop is already powered on.")

    def shutdown(self):
        if self.powered_on:
            print(f"{self.brand} {self.model} desktop is shutting down.")
            self.powered_on = False
        else:
            print("Desktop is already powered off.")

class Laptop(ComputerSystem):
    def power_on(self):
        if not self.powered_on:
            print(f"{self.brand} {self.model} laptop is powering on.")
            self.powered_on = True
        else:
            print("Laptop is already powered on.")

    def shutdown(self):
        if self.powered_on:
            print(f"{self.brand} {self.model} laptop is shutting down.")
            self.powered_on = False
        else:
            print("Laptop is already powered off.")

# Example usage:
desktop = Desktop(brand="Dell", model="OptiPlex")
laptop = Laptop(brand="HP", model="EliteBook")

systems = [desktop, laptop]

for system in systems:
    system.power_on()
    system.shutdown()
    system.shutdown()
```

```python
# 19. Create a Python class for a library system, implementing abstraction by defining common methods (e.g., `add_book()`, `borrow_book()`) in an abstract base class.
from abc import ABC, abstractmethod

class LibrarySystem(ABC):
    def __init__(self, name):
        self.name = name
        self.books = {}

    @abstractmethod
    def add_book(self, title, author):
        pass

    @abstractmethod
    def borrow_book(self, title):
        pass

    @abstractmethod
    def return_book(self, title):
        pass

class SimpleLibrary(LibrarySystem):
    def add_book(self, title, author, copies=1):
        if title in self.books:
            self.books[title]['copies'] += copies
        else:
            self.books[title] = {'author': author, 'copies': copies}

        print(f"Book '{title}' by {author} added to the library.")

    def borrow_book(self, title):
        if title in self.books and self.books[title]['copies'] > 0:
            self.books[title]['copies'] -= 1
            print(f"Book '{title}' borrowed successfully.")
        else:
            print(f"Sorry, '{title}' is not available for borrowing.")

    def return_book(self, title):
        if title in self.books:
            self.books[title]['copies'] += 1
            print(f"Book '{title}' returned.")
        else:
            print(f"Cannot return '{title}' as it is not part of the library.")

# Example usage:
library = SimpleLibrary(name="My Library")

library.add_book(title="The Catcher in the Rye", author="J.D. Salinger", copies=5)
library.add_book(title="To Kill a Mockingbird", author="Harper Lee", copies=3)

library.borrow_book(title="The Catcher in the Rye")
library.borrow_book(title="To Kill a Mockingbird")

library.return_book(title="The Catcher in the Rye")
library.return_book(title="Nonexistent Book")

# Output:
# Book 'The Catcher in the Rye' by J.D. Salinger added to the library.
# Book 'To Kill a Mockingbird' by Harper Lee added to the library.
# Book 'The Catcher in the Rye' borrowed successfully.
# Book 'To Kill a Mockingbird' borrowed successfully.
# Book 'The Catcher in the Rye' returned.
# Cannot return 'Nonexistent Book' as it is not part of the library.
```

```python
# 5. How can you implement composition in Python classes? Provide examples of using composition to create complex objects.
# Example 1: Basic Composition

class Engine:
    def start(self):
        return "Engine started"

class Wheels:
    def rotate(self):
        return "Wheels rotating"

class Car:
    def __init__(self):
        self.engine = Engine()  # Engine is a component of Car
        self.wheels = Wheels()  # Wheels is a component of Car

    def drive(self):
        return f"{self.engine.start()} and {self.wheels.rotate()}"

# Creating a Car object using composition
my_car = Car()
print(my_car.drive())
```

```python
# Example 2: More Complex Composition

class CPU:
    def process_data(self):
        return "Data processing by CPU"

class Memory:
    def store_data(self):
        return "Data stored in memory"

class GPU:
    def render_graphics(self):
        return "Graphics rendered by GPU"

class Computer:
    def __init__(self):
        self.cpu = CPU()      # CPU is a component of Computer
        self.memory = Memory()  # Memory is a component of Computer
        self.gpu = GPU()      # GPU is a component of Computer

    def perform_tasks(self):
        return f"{self.cpu.process_data()}, {self.memory.store_data()}, and {self.gpu.render_graphics()}"

# Creating a Computer object using composition
my_computer = Computer()
print(my_computer.perform_tasks())
```

```python
# 6. Create a Python class hierarchy for a music player system, using composition to represent playlists and songs.
class Song:
    def __init__(self, title, artist, duration):
        self.title = title
        self.artist = artist
        self.duration = duration

    def play(self):
        return f"Now playing: {self.title} by {self.artist}"

class Playlist:
    def __init__(self, name):
        self.name = name
        self.songs = []

    def add_song(self, song):
        self.songs.append(song)

    def play_all(self):
        if not self.songs:
            return "Playlist is empty."
        else:
            return "\n".join(song.play() for song in self.songs)

class MusicPlayer:
    def __init__(self):
        self.playlists = []

    def add_playlist(self, playlist):
        self.playlists.append(playlist)

    def play_all_playlists(self):
        if not self.playlists:
            return "No playlists available."
        else:
            return "\n\n".join(f"Playlist: {playlist.name}\n{playlist.play_all()}" for playlist in self.playlists)

# Example usage:
song1 = Song(title="Imagine", artist="John Lennon", duration="3:00")
song2 = Song(title="Bohemian Rhapsody", artist="Queen", duration="5:55")
song3 = Song(title="Shape of You", artist="Ed Sheeran", duration="4:23")

playlist1 = Playlist(name="Favorites")
playlist1.add_song(song1)
playlist1.add_song(song2)

playlist2 = Playlist(name="Party Mix")
playlist2.add_song(song2)
playlist2.add_song(song3)

music_player = MusicPlayer()
music_player.add_playlist(playlist1)
music_player.add_playlist(playlist2)

print(music_player.play_all_playlists())
```

```python
# 7. Explain the concept of "has-a" relationships in composition and how it helps design software systems.
# The concept of "has-a" relationships in composition refers to the idea that one class contains an instance of another class as an attribute. This type of relationship is fundamental to composition, and it helps design software systems by allowing them to be built from smaller, reusable components. The "has-a" relationship signifies that an object of one class possesses another object, indicating a composition of objects to create a more complex system.

# Here are key points explaining the concept of "has-a" relationships in composition:

# 1. **Component-Based Design:**
 #   - In a "has-a" relationship, a class is designed to have components or parts that contribute to its functionality.
 #   - Components are represented by instances of other classes, encapsulating specific behaviors or properties.

# 2. **Modularity and Reusability:**
 #  - The "has-a" relationship promotes modularity by breaking down a system into smaller, self-contained components.
  #  - Each class is responsible for a specific aspect of functionality, enhancing code organization and making it more modular.
  #  - Components can be reused in different contexts, leading to greater code reusability.

# 3. **Encapsulation:**
  #  - Each class involved in a "has-a" relationship encapsulates its own functionality.
  #  - Changes to the implementation of one class do not directly affect other classes, promoting encapsulation and reducing the risk of unintended consequences.

# 4. **Flexibility and Adaptability:**
  #  - The "has-a" relationship allows for dynamic composition at runtime.
   # - Objects can be assembled, modified, or replaced during the execution of the program, providing flexibility and adaptability to changing requirements.

# 5. **Clear Representation of Real-World Relationships:**
  #  - The "has-a" relationship often mirrors real-world relationships more accurately than an "is-a" relationship (inheritance).
   # - For example, a `Car` "has-a" `Engine` and "has-a" set of `Wheels`, reflecting a more natural and intuitive representation.

# 6. **Avoidance of Deep Class Hierarchies:**
   # - Composition helps avoid issues associated with deep class hierarchies, such as the "diamond problem" in multiple inheritance.
  #  - Instead of creating a complex hierarchy, classes are composed of other classes, reducing complexity and potential conflicts.

```

```python
# 8. Create a Python class for a computer system, using composition to represent components like CPU, RAM, and storage devices.
class CPU:
    def __init__(self, brand, speed):
        self.brand = brand
        self.speed = speed

    def process_data(self):
        return f"{self.brand} CPU processing data at {self.speed} GHz"

class RAM:
    def __init__(self, capacity):
        self.capacity = capacity

    def load_data(self):
        return f"RAM with {self.capacity} GB capacity loading data"

class StorageDevice:
    def __init__(self, type, capacity):
        self.type = type
        self.capacity = capacity

    def read_data(self):
        return f"{self.type} storage device reading data, capacity: {self.capacity} GB"

class ComputerSystem:
    def __init__(self, cpu, ram, storage):
        self.cpu = cpu  # Composition: ComputerSystem has-a CPU
        self.ram = ram  # Composition: ComputerSystem has-a RAM
        self.storage = storage  # Composition: ComputerSystem has-a StorageDevice

    def run_program(self, program):
        cpu_result = self.cpu.process_data()
        ram_result = self.ram.load_data()
        storage_result = self.storage.read_data()
        return f"{cpu_result}\n{ram_result}\n{storage_result}\nRunning program: {program}"

# Creating instances for components
my_cpu = CPU(brand="Intel", speed=2.5)
my_ram = RAM(capacity=8)
my_storage = StorageDevice(type="SSD", capacity=256)

# Creating a ComputerSystem with composition
my_computer = ComputerSystem(cpu=my_cpu, ram=my_ram, storage=my_storage)

# Running a program on the computer system
result = my_computer.run_program(program="Example Program")

# Displaying the result
print(result)
```

```python
# 9. Describe the concept of "delegation" in composition and how it simplifies the design of complex systems.
# Delegation in composition is a design concept where one object passes on a task to another object by invoking its methods. Instead of inheriting behavior, an object delegates certain responsibilities or functionalities to another object, which is typically composed as part of the first object. This approach simplifies the design of complex systems by promoting modularity, encapsulation, and flexibility.

# Key aspects of delegation in composition:

# 1. **Modularity:**
 #   - Delegation encourages a modular design where each class is responsible for a specific aspect of functionality.
  #  - Objects can delegate tasks to other objects, creating a clear separation of concerns.

# 2. **Encapsulation:**
  #  - Each object involved in delegation encapsulates its own functionality.
   # - The internal details of how a task is accomplished are hidden from the delegating object, promoting encapsulation.

# 3. **Loose Coupling:**
  #  - Delegation results in loose coupling between objects. The delegating object is not dependent on the internal implementation details of the delegated object.
  #  - Changes to the delegated object's implementation do not affect the delegating object, fostering a more resilient and maintainable system.

# 4. **Flexibility:**
  #  - Delegation provides flexibility in terms of dynamically changing the behavior of an object.
  #  - Objects can be swapped or replaced with different ones that implement the delegated functionality, allowing for runtime adaptability.

# 5. **Reusability:**
  #  - Delegation promotes code reusability by allowing objects to reuse the functionality provided by other objects without inheritance.
  #  - Different objects can delegate tasks to a common set of objects, facilitating a more flexible and reusable design.

# 6. **Clear Responsibilities:**
   # - Delegation helps assign clear responsibilities to objects in a system.
  #  - Each object focuses on its specific area of expertise, and delegation allows tasks to be distributed among objects based on their capabilities.

# 7. **Dynamic Behavior:**
  #  - Delegation enables dynamic behavior, where the behavior of an object can be modified during runtime by changing the delegated object.
  #  - This dynamic nature is beneficial in scenarios where different behaviors are required under different conditions.

# 8. **Simplified Maintenance:**
 #   - Delegation often leads to simplified maintenance, as changes to one part of the system are less likely to have a ripple effect on other parts.
 #   - Modifying or extending functionality can be done in a localized manner, reducing the risk of unintended side effects.

```

```python
# 10. Create a Python class for a car, using composition to represent components like the engine, wheels, and transmission.
class Engine:
    def __init__(self, fuel_type, horsepower):
        self.fuel_type = fuel_type
        self.horsepower = horsepower

    def start(self):
        return "Engine started"

    def stop(self):
        return "Engine stopped"

class Wheels:
    def __init__(self, brand):
        self.brand = brand

    def rotate(self):
        return "Wheels rotating"

class Transmission:
    def __init__(self, transmission_type):
        self.transmission_type = transmission_type

    def shift_gear(self):
        return "Gear shifted"

class Car:
    def __init__(self, engine, wheels, transmission):
        self.engine = engine  # Composition: Car has-an Engine
        self.wheels = wheels  # Composition: Car has-Wheels
        self.transmission = transmission  # Composition: Car has-a Transmission

    def start(self):
        engine_result = self.engine.start()
        wheels_result = self.wheels.rotate()
        transmission_result = self.transmission.shift_gear()
        return f"{engine_result}\n{wheels_result}\n{transmission_result}\nCar started"

    def stop(self):
        engine_result = self.engine.stop()
        return f"{engine_result}\nCar stopped"

# Creating instances for components
car_engine = Engine(fuel_type="Gasoline", horsepower=200)
car_wheels = Wheels(brand="Michelin")
car_transmission = Transmission(transmission_type="Automatic")

# Creating a Car with composition
my_car = Car(engine=car_engine, wheels=car_wheels, transmission=car_transmission)

# Starting and stopping the car
start_result = my_car.start()
print(start_result)

stop_result = my_car.stop()
print(stop_result)
```

```python
# 11. How can you encapsulate and hide the details of composed objects in Python classes to maintain abstraction?
# In Python, you can encapsulate and hide the details of composed objects in classes to maintain abstraction by following certain principles. Here are some techniques:

# Private Attributes:

# Use private attributes (prefixed with a double underscore __) to hide the details of composed objects within a class.
# This makes the attributes inaccessible from outside the class, providing a level of encapsulation.
class Car:
    def __init__(self, engine, wheels, transmission):
        self.__engine = engine
        self.__wheels = wheels
        self.__transmission = transmission

    def start(self):
        engine_result = self.__engine.start()
        wheels_result = self.__wheels.rotate()
        transmission_result = self.__transmission.shift_gear()
        return f"{engine_result}\n{wheels_result}\n{transmission_result}\nCar started"

# Creating instances for components
car_engine = Engine(fuel_type="Gasoline", horsepower=200)
car_wheels = Wheels(brand="Michelin")
car_transmission = Transmission(transmission_type="Automatic")

# Creating a Car with composition
my_car = Car(engine=car_engine, wheels=car_wheels, transmission=car_transmission)

# Trying to access private attributes directly will result in an AttributeError
# print(my_car.__engine)  # Raises AttributeError
```

```python
# Getter Methods:

# Provide getter methods to allow controlled access to the details of composed objects.
# This allows external code to access certain information in a controlled manner, maintaining encapsulation.
class Car:
    def __init__(self, engine, wheels, transmission):
        self.__engine = engine
        self.__wheels = wheels
        self.__transmission = transmission

    def get_engine_info(self):
        return f"Engine: {self.__engine.fuel_type}, {self.__engine.horsepower} HP"

    def get_wheels_info(self):
        return f"Wheels: {self.__wheels.brand}"

    def get_transmission_info(self):
        return f"Transmission: {self.__transmission.transmission_type}"

# Creating instances for components
car_engine = Engine(fuel_type="Gasoline", horsepower=200)
car_wheels = Wheels(brand="Michelin")
car_transmission = Transmission(transmission_type="Automatic")

# Creating a Car with composition
my_car = Car(engine=car_engine, wheels=car_wheels, transmission=car_transmission)

# Accessing details through getter methods
print(my_car.get_engine_info())
print(my_car.get_wheels_info())
print(my_car.get_transmission_info())
```

```python
# Abstraction through Interfaces or Abstract Classes:

# Define interfaces or abstract classes for the composed objects, and use them in the containing class.
# This allows you to interact with the composed objects through well-defined interfaces, hiding their implementation details.
from abc import ABC, abstractmethod

class Engine(ABC):
    @abstractmethod
    def start(self):
        pass

class Wheels(ABC):
    @abstractmethod
    def rotate(self):
        pass

class Transmission(ABC):
    @abstractmethod
    def shift_gear(self):
        pass

class Car:
    def __init__(self, engine: Engine, wheels: Wheels, transmission: Transmission):
        self.__engine = engine
        self.__wheels = wheels
        self.__transmission = transmission

    def start(self):
        engine_result = self.__engine.start()
        wheels_result = self.__wheels.rotate()
        transmission_result = self.__transmission.shift_gear()
        return f"{engine_result}\n{wheels_result}\n{transmission_result}\nCar started"

# Implementing concrete classes for Engine, Wheels, and Transmission
class GasolineEngine(Engine):
    def start(self):
        return "Gasoline engine started"

class MichelinWheels(Wheels):
    def rotate(self):
        return "Michelin wheels rotating"

class AutomaticTransmission(Transmission):
    def shift_gear(self):
        return "Automatic transmission shifted gear"

# Creating instances for components
car_engine = GasolineEngine()
car_wheels = MichelinWheels()
car_transmission = AutomaticTransmission()

# Creating a Car with composition
my_car = Car(engine=car_engine, wheels=car_wheels, transmission=car_transmission)

# Starting the car
print(my_car.start())
```

```python
# 12. Create a Python class for a university course, using composition to represent students, instructors, and course materials.
class Student:
    def __init__(self, student_id, name):
        self.student_id = student_id
        self.name = name

    def enroll(self, course):
        return f"{self.name} enrolled in {course.course_name}"

class Instructor:
    def __init__(self, instructor_id, name):
        self.instructor_id = instructor_id
        self.name = name

    def teach(self, course):
        return f"{self.name} is teaching {course.course_name}"

class CourseMaterial:
    def __init__(self, textbooks, syllabus):
        self.textbooks = textbooks
        self.syllabus = syllabus

class UniversityCourse:
    def __init__(self, course_code, course_name, instructor, students, course_material):
        self.course_code = course_code
        self.course_name = course_name
        self.instructor = instructor  # Composition: UniversityCourse has-an Instructor
        self.students = students  # Composition: UniversityCourse has-Students
        self.course_material = course_material  # Composition: UniversityCourse has-a CourseMaterial

    def display_course_info(self):
        instructor_info = self.instructor.teach(self)
        students_info = "\n".join(student.enroll(self) for student in self.students)
        return f"Course Code: {self.course_code}\nCourse Name: {self.course_name}\n{instructor_info}\n{students_info}"

# Creating instances for students
student1 = Student(student_id="S001", name="Alice")
student2 = Student(student_id="S002", name="Bob")

# Creating an instance for an instructor
instructor = Instructor(instructor_id="I001", name="Dr. Smith")

# Creating an instance for course materials
textbooks = ["Textbook A", "Textbook B"]
syllabus = "Course Syllabus for Computer Science"

course_material = CourseMaterial(textbooks=textbooks, syllabus=syllabus)

# Creating a UniversityCourse with composition
computer_science_course = UniversityCourse(
    course_code="CS101",
    course_name="Introduction to Computer Science",
    instructor=instructor,
    students=[student1, student2],
    course_material=course_material
)

# Displaying course information
course_info = computer_science_course.display_course_info()
print(course_info)
```

```python
# 13. Discuss the challenges and drawbacks of composition, such as increased complexity and potential for tight coupling between objects.
# While composition offers many benefits, it comes with its own set of challenges and drawbacks. Understanding these drawbacks is crucial for making informed design decisions. Here are some challenges associated with composition:

# 1. **Increased Complexity:**
 #   - As the number of composed objects increases, the overall complexity of the system may rise.
 #   - Managing relationships and interactions between multiple objects can become challenging, especially in large systems.

# 2. **Boilerplate Code:**
  #  - Composition may lead to the creation of additional boilerplate code, especially when forwarding methods or implementing interfaces for composed objects.
  #  - This can result in more code to write and maintain, potentially leading to verbosity.

# 3. **Potential for Tight Coupling:**
  #  - While composition aims to reduce tight coupling compared to inheritance, improper use can still lead to dependencies between objects.
 #   - Changes in one component may necessitate changes in dependent components, leading to a degree of coupling.

# 4. **Interface Proliferation:**
  #  - In an attempt to provide clear interfaces for composed objects, there might be a tendency to create numerous small interfaces or abstract classes.
  #  - This can lead to interface proliferation, making the system harder to understand and navigate.

# 5. **Runtime Overhead:**
 #   - Composition may introduce runtime overhead due to the need for dynamic delegation or forwarding of method calls.
 #   - This overhead can impact performance in certain scenarios, especially in resource-intensive applications.

# 6. **Difficulties in Object Creation:**
  #  - Creating instances of composed objects and configuring them appropriately can be more complex than inheriting from a base class.
  #  - Object creation and initialization may require more explicit code, impacting readability.

```

```python
# 14. Create a Python class hierarchy for a restaurant system, using composition to represent menus, dishes, and ingredients.
class Ingredient:
    def __init__(self, name, quantity, unit):
        self.name = name
        self.quantity = quantity
        self.unit = unit

class Dish:
    def __init__(self, name, description, ingredients):
        self.name = name
        self.description = description
        self.ingredients = ingredients  # Composition: Dish has-Ingredients

    def display_dish_info(self):
        ingredient_info = "\n".join(f"{ingredient.quantity} {ingredient.unit} of {ingredient.name}" for ingredient in self.ingredients)
        return f"Dish: {self.name}\nDescription: {self.description}\nIngredients:\n{ingredient_info}"

class Menu:
    def __init__(self, name, dishes):
        self.name = name
        self.dishes = dishes  # Composition: Menu has-Dishes

    def display_menu_info(self):
        dish_info = "\n\n".join(dish.display_dish_info() for dish in self.dishes)
        return f"Menu: {self.name}\n\n{dish_info}"

# Creating instances for ingredients
ingredient1 = Ingredient(name="Tomato", quantity=2, unit="pieces")
ingredient2 = Ingredient(name="Cheese", quantity=100, unit="grams")
ingredient3 = Ingredient(name="Flour", quantity=200, unit="grams")

# Creating instances for dishes
dish1 = Dish(name="Margherita Pizza", description="Classic pizza with tomato and cheese", ingredients=[ingredient1, ingredient2, ingredient3])
dish2 = Dish(name="Pasta Carbonara", description="Creamy pasta with bacon and cheese", ingredients=[ingredient2, ingredient3])

# Creating an instance for a menu
restaurant_menu = Menu(name="Italian Delight", dishes=[dish1, dish2])

# Displaying menu information
menu_info = restaurant_menu.display_menu_info()
print(menu_info)
```

```python
# 18. Create a Python class for a house, using composition to represent rooms, furniture, and appliances.
class Furniture:
    def __init__(self, name):
        self.name = name

    def describe(self):
        return f"A piece of furniture: {self.name}"

class Appliance:
    def __init__(self, name):
        self.name = name

    def operate(self):
        return f"The {self.name} is now operating."

class Room:
    def __init__(self, name, furniture=None, appliances=None):
        self.name = name
        self.furniture = furniture if furniture else []
        self.appliances = appliances if appliances else []

    def describe(self):
        furniture_info = "\n".join(furniture.describe() for furniture in self.furniture)
        appliances_info = "\n".join(appliance.operate() for appliance in self.appliances)
        return f"Room: {self.name}\nFurniture:\n{furniture_info}\nAppliances:\n{appliances_info}"

class House:
    def __init__(self, rooms):
        self.rooms = rooms  # Composition: House has-Rooms

    def describe_house(self):
        room_info = "\n\n".join(room.describe() for room in self.rooms)
        return f"House Description:\n{room_info}"

# Creating instances for furniture and appliances
sofa = Furniture(name="Sofa")
table = Furniture(name="Coffee Table")
oven = Appliance(name="Oven")
tv = Appliance(name="Smart TV")

# Creating instances for rooms
living_room = Room(name="Living Room", furniture=[sofa, table], appliances=[tv])
kitchen = Room(name="Kitchen", appliances=[oven])

# Creating a House with composition
my_house = House(rooms=[living_room, kitchen])

# Displaying house information
house_info = my_house.describe_house()
print(house_info)
```

```python
# 19. How can you achieve flexibility in composed objects by allowing them to be replaced or modified dynamically at runtime?
# Achieving flexibility in composed objects, allowing them to be replaced or modified dynamically at runtime, involves designing the system to be adaptable and open to changes. Here are several techniques to achieve this flexibility:

# 1. **Interfaces and Abstract Classes:**
 #   - Define interfaces or abstract classes for components to establish a common contract.
  #  - Allow different implementations of components to adhere to the same interface, making them interchangeable.
  #  - Use polymorphism to dynamically replace objects with others that share the same interface.

# 2. **Dependency Injection:**
 #   - Inject dependencies into objects rather than hard-coding them.
 #   - Allow objects to be created with their dependencies passed as parameters, making it easy to replace or modify components at runtime.

# 3. **Factory Patterns:**
  #  - Use factory patterns to create objects dynamically.
   # - Factories encapsulate the logic of object creation, allowing for the instantiation of different types of objects based on certain conditions.

# 4. **Configuration Files:**
   # - Store configuration details, such as class names or types, in external configuration files.
  #  - Dynamically read and instantiate objects based on the configuration, providing a way to modify the composition without changing the code.

# 5. **Dependency Inversion Principle (DIP):**
  #  - Apply the Dependency Inversion Principle, which suggests that high-level modules should not depend on low-level modules, but both should depend on abstractions.
  #  - Depend on abstractions rather than concrete implementations, making it easier to replace components with different implementations.

# 6. **Service Locator Pattern:**
   # - Use a service locator to centralize the creation and location of services or components.
  #  - This allows for dynamic replacement of services by updating the service locator.

```

```python
# 20. Create a Python class for a social media application, using composition to represent users, posts, and comments.
class Comment:
    def __init__(self, user, text):
        self.user = user
        self.text = text

    def display(self):
        return f"{self.user.username}: {self.text}"

class Post:
    def __init__(self, user, content):
        self.user = user
        self.content = content
        self.comments = []  # Composition: Post has-Comments

    def add_comment(self, comment):
        self.comments.append(comment)

    def display_post(self):
        post_info = f"{self.user.username} posted:\n{self.content}"
        comment_info = "\nComments:\n" + "\n".join(comment.display() for comment in self.comments)
        return post_info + comment_info

class User:
    def __init__(self, username):
        self.username = username
        self.posts = []  # Composition: User has-Posts

    def create_post(self, content):
        post = Post(user=self, content=content)
        self.posts.append(post)
        return post

# Creating instances for users, posts, and comments
user1 = User(username="Alice")
user2 = User(username="Bob")

post1 = user1.create_post(content="Hello, world!")
comment1 = Comment(user=user2, text="Nice post!")

post1.add_comment(comment1)

# Displaying social media information
post_info = post1.display_post()
print(post_info)
```
